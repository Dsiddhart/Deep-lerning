```python
# key-value, unordered, mutable
daict={"name":"max","age":24,"city":"vns"}
print(daict)
```

```python
dict2=__builtins__.dict(name="ggg",age=23,)
print(dict2)
```

```python
dict2["city"]="vns"
print(dict2)
```

```python
del dict2["name"]
print(dict2)
```

```python
dict2.pop("city")
print(dict2)
```

```python
daict.popitem()
print(daict)
```

```python
try:
  print(daict["city"])
except:
  print("error")
```

```python
for key in daict:
  print(key)
```

```python
for key,value in daict.items():
  print(key,value)
```

```python
dicti=daict
dicti["name"]="ggg"
print(dicti)
print(daict)
```

```python
dcp=dicti.copy()
dcp["email"]="dwfe@efef.com"
print(dicti)
print(dcp)
```

```python
daict.update(dcp)
print(daict)
```
