```python
#Tuple- orderer, immutable, allow duplication
```

```python
mytuple=("Max",23,True,24,2)
print(mytuple)
print(mytuple[2])
```

```python
tup="max",22,4,True
print(tup)
```

```python
tupi="max"
print(type(tupi))
```

```python
tupi="max",
print(type(tupi))
```

```python
mtup=tuple(["jj",44,2,True])
print(mtup)
print(mtup[-1])
```

```python
mtup[0]="aa"
```

```python
for x in mtup:
  print(x)
```

```python
if "jj " in mtup:
  print("yes")
else:
  print("no")
```

```python
ntp=('a','v','d','e','g')
print(len(ntp))
print(ntp.count('a'))
print(ntp.index('d'))
ml=list(ntp)
print(ml)
mtp=tuple(ml)
print(mtp)
```

```python
nt=(1,2,3,4,5,6,7)
b=nt[:5]
print(b)
```

```python
nt=(1,2,3,4,5,6,7)
b=nt[::-1]
print(b)
```

```python
tup="max",22,4,True
name,age,car,is_alive=tup
print(name)
print(age)
print(car)
print(is_alive)
#unpacking
```

```python
nt=(1,2,3,4,5,6,7)
i1,*i2,i3=nt
print(i3)
print(i1)
print(*i2)
print(type(i2))
```

```python
import sys
my_list=['a', 'v', 'd', 'e', 'g']
my_tup=('a','v','d','e','g')
print(sys.getsizeof(my_list,'bytes'))
print(sys.getsizeof(my_tup,'bytes'))
```

```python
import timeit
print(timeit.timeit(stmt="[0,1,2,3,4,5]",number=1000000))
print(timeit.timeit(stmt="(0,1,2,3,4,5)",number=1000000))
```
