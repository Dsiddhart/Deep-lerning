```python
import math

def sigmoid(x):
  y=1.0/(1+ math.exp(-x))
  return y

def activate(inputs,weights):
  h=0
  for x,w in zip(inputs,weights):
    h+=x*w

  return sigmoid(h)



if __name__ == "__main__":
  inputs=[.5,.3,.2]#list
  weights=[.4,.7,.2]
  output=activate(inputs,weights)
  print(output)
```

**Implementing a neural network**


```python
import numpy as np

class MLP:

  def __init__(self,num_input=3,num_hlayer=[3,5],num_output=2):

    self.num_input=num_input;
    self.num_hlayer=num_hlayer;
    self.num_output=num_output;
    # no. of neurons in each layers
    layers_neurons=[self.num_input]+self.num_hlayer+[self.num_output];

    # initiate random weights

    self.weights=[]
    for i in range(len(layers_neurons)-1):
      w=np.random.rand(layers_neurons[i],layers_neurons[i+1])# (0,1)
      self.weights.append(w)

  def forward_propagate(self,inputs):
    activations=inputs
    print("no.of inputs in starting:{}".format(activations))
    for w in self.weights:
      # calculate net input
      print("no.of weights:{}".format(w))
      net_inputs=np.dot(activations,w)

      print("no.of net inputs:{}".format(net_inputs))

      # calculate the activations

      activations =self._sigmoid(net_inputs)
      print("no.of inputs after activation function:{}".format(activations))
    return activations

  def _sigmoid(self,x):
    return 1/(1+np.exp(-x))


if __name__=="__main__":

  # create an mlp
  mlp=MLP()
  #create some inputs
  inputs= np.random.rand(mlp.num_input)
  # perform forward prop

  outputs=mlp.forward_propagate(inputs)

  # print the result
  print("the network input is:{}".format(inputs))
  print("the network output is:{}".format(outputs))




```
